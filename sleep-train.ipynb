```python
# X_train.shape

import numpy as np

def sequentialize(features,labels,seq_len):
    seq_features = []
    seq_labels = []
    for i in range(len(features) - seq_len):
        seq_features.append(features[i:i + seq_len])
        seq_labels.append(labels[i+seq_len])

    seq_features = np.array(seq_features)
    seq_labels = np.array(seq_labels)

    test_size = int(np.round(0.3 * seq_features.shape[0]))
    train_size = seq_features.shape[0] - test_size
    
    print(seq_features.shape)
    print(seq_labels.shape)
    X_train = seq_features[:train_size,:,: ]
    y_train = seq_labels[:train_size,:]
    X_test = seq_features[train_size:,:,:]
    y_test = seq_labels[train_size:,:]


    return X_train, y_train, X_test, y_test


X_train,y_train,X_test, y_test= sequentialize(features, labels, 40)
```

```python
X_train = np.transpose(X_train, (0, 2, 1))
X_test = np.transpose(X_test, (0, 2, 1))
```

```python
# Define your custom dataset
from torch.utils.data import Dataset, DataLoader

class SleepDataset(Dataset):
    def __init__(self, sequences, target):
        self.sequences = torch.from_numpy(sequences).float()
        self.target = torch.from_numpy(target).float()

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        sequence = self.sequences[idx]
        target = self.target[idx]

        return {
            'sequence': sequence,
            'target': target
        }
```

```python
# Create custom datasets and dataloaders
train_dataset = SleepDataset(X_train, y_train)
test_dataset = SleepDataset(X_test, y_test)
batch_size = 32
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False,drop_last=True)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False,drop_last =True)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics import precision_score, recall_score, f1_score
from torch.nn import MultiheadAttention


class SleepModel(nn.Module):
    def __init__(self, feat_dim, embed_dim, num_heads, ff_dim):
        super(SleepModel, self).__init__()

        # CNN LAYERS
        self.conv1 = nn.Conv1d(feat_dim, 64, kernel_size=5, stride=3, padding=1)
        self.conv2 = nn.Conv1d(64, 64, kernel_size=3, stride=2, padding=1)
        self.conv3 = nn.Conv1d(64, 64, kernel_size=3, stride=2, padding=1)

        # SIX RNN BLOCKS
        self.gru_blocks = nn.ModuleList([nn.GRU(4, 16, bidirectional=True, batch_first=True)] + 
                                [nn.GRU(32, 16, bidirectional=True, batch_first=True) for _ in range(5)])

        self.attn = MultiheadAttention(embed_dim=64, num_heads=num_heads)  

        # Fully Connected Layer
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64,1)
        )
        
    def forward(self, x):
        # CNN LAYERS
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        # SIX RNN BLOCKS
        for gru_block in self.gru_blocks:
            x, _ = gru_block(x)
        
    
        x = x[:,:,-1]

        x, _ = self.attn(x, x, x)  
        x = x.flatten(start_dim=1)

        x = self.fc(x)
        return x


feat_dim = 2  
embed_dim = 32  
num_heads = 4  
ff_dim = 128  
model = SleepModel(feat_dim, embed_dim, num_heads, ff_dim).to(device)
```

```python
criterion = nn.BCEWithLogitsLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
def accuracy_fn(y_true,y_preds):
    correct = torch.eq(y_true,y_preds).sum().item()
    acc = correct/len(y_true)
    return acc*100
```

```python
# Training and Testing Loop
from sklearn.metrics import accuracy_score

from tqdm import tqdm
num_epochs = 10

for epoch in tqdm(range(num_epochs)):
    avg_loss= 0
    total_acc = 0
    for batch_idx,batch in enumerate(train_dataloader):
        model.train()
        
        input_sequences, target = batch['sequence'].to(device), batch['target'].to(device)
        # Forward pass
        output = model(input_sequences).to(device)   
        mask = ~torch.isnan(target)
        
        predictions = torch.round(torch.sigmoid(output)) 
        loss = criterion(output[mask], target[mask])
        if(torch.isnan(loss)):
            g=0
        else:
            optimizer.zero_grad()

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            avg_loss+=loss.item()
        
    avg_loss = avg_loss / len(train_dataloader)
    print(f'Epoch [{epoch+1}/{num_epochs}], '
          f'Loss: {avg_loss}') 
```

```python
def ensemble_predictions(models, dataloader):
    predictions = []
    for model in models:
        model.eval()
        with torch.no_grad():
            model_predictions = []
            for inputs, _ in dataloader:
                outputs = model(inputs)
                outputs = torch.argmax(outputs,dim=1)
                model_predictions.extend(outputs.numpy())

            predictions.append(model_predictions)

    # Combine predictions by averaging (you can also experiment with voting or other aggregation methods)
    ensemble_predictions = torch.tensor(np.mean(predictions, axis=0), dtype=torch.float32)
    return ensemble_predictions

# ensemble_val_predictions = ensemble_predictions(model, test_dataloader)
```
